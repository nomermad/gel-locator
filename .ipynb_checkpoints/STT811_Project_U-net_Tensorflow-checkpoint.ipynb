## Res-U-net

-------
The res-u-net is the backbone of the models used for the project. Res-u-net differs in a number of ways from base u-net, but they both function the same way.

```python
# Res-U-net conv block
def batchnorm_relu(inputs):
     """ Batch Normalization & ReLU """
     x = BatchNormalization()(inputs)
     x = Activation("relu")(x)
     return x

def residual_block(inputs, num_filters, strides=1):
    """ Convolutional Layers """
    x = batchnorm_relu(inputs)
    x = Conv2D(num_filters, 3, padding="same", strides=strides)(x)
    x = batchnorm_relu(x)
    x = Conv2D(num_filters, 3, padding="same", strides=1)(x)
    """ Shortcut Connection (Identity Mapping) """
    s = Conv2D(num_filters, 1, padding="same", strides=strides)(inputs)
    """ Addition """
    x = x + s
    return x

def decoder_block(inputs, skip_features, num_filters):
    """ Decoder Block """
    x = UpSampling2D((2, 2))(inputs) 
    x = Concatenate()([x, skip_features]) 
    x = residual_block(x, num_filters, strides=1) 
    return x

def build_resunet(input_shape):
    """ RESUNET Architecture """
    
    inputs = Input(input_shape)

    """ Endoder 1 """ 
    x = Conv2D(64, 3, padding="same", strides=1)(inputs) 
    x = batchnorm_relu(x) 
    x = Conv2D(64, 3, padding="same", strides=1)(x) 
    s = Conv2D(64, 1, padding="same")(inputs) 
    s1 = x + s

    """ Encoder 2, 3 """
    s2 = residual_block(s1, 128, strides=2)
    s3 = residual_block(s2, 256, strides=2)

    """ Bridge """
    b = residual_block(s3, 512, strides=2)

    """ Decoder 1, 2, 3 """ 
    x = decoder_block(b, s3, 256) 
    x = decoder_block(x, s2, 128) 
    x = decoder_block(x, s1, 64)

    """ Classifier """
    outputs = Conv2D(1, 1, padding="same", activation="sigmoid")(x)

    """ Model """
    model = Model(inputs, outputs, name="RESUNET")

    return model
```

```python
model = build_resunet(input_shape)
# Types of loss functions
# Binary crossentropy: 'binary_crossentropy'
# Binary Focal Crossentropy: 'binary_focal_crossentropy'
model.compile(optimizer=Adam(learning_rate = 1e-3), loss="binary_focal_crossentropy", metrics=['accuracy']) # 'accuracy' might be a poor metric to use as it fails to incorporate what we really want to have classified.
model.summary()
```

--------
Given how expensive it is to run a neural network, its best to recycle model wights and use them for new models. Transferable learning is the boosting of neural networks where a model that predicts data close to the training data is loaded, the part of the model that identify impotent features is frozen, and the model only adapts to refining classification bounds.
